```python
import numpy as np
from numpy import random as random
import matplotlib.pyplot as plt
```

```python
nSamples = 10000
x = random.rand(nSamples) # Get random samples from a uniform distribution
#plt.stem(x)
#plt.show()

a = np.arange(0,1,0.01)
CDF_x = np.zeros(a.size)
for ia in range(a.size):
    CDF_x[ia] = np.count_nonzero(x<=a[ia])/nSamples

plt.plot(a,CDF_x)
plt.show()

y = x**2 # y is another random variable dependent on x

CDF_y = np.zeros(a.size)
for ia in range(a.size):
    CDF_y[ia] = np.count_nonzero(y<=a[ia])/nSamples

plt.plot(a,CDF_y)
#plt.plot(a,np.sqrt(a))
plt.show()
```

The CFD of $y$ is *not* uniform, because:

$X~UD[0,1]$

$Y=X^2$

CDF of Y is given by:

$F_Y(y)=P(Y\le y)=P(X^2\le y)$

$=P(X \le \sqrt{y})$

$=\sqrt{y}$

```python
# Get and plot the PDF of y
counts, bins_edge = np.histogram(y,bins=20,density=True)
plt.plot(bins_edge[1:],counts)
plt.show()
```

$\mu_Y = E[Y] = \int_{0}^{1} y \text{PDF}(y) dy=\frac{1}{3}$

```python
mu_y = sum(y)/len(y)
print(mu_y)
```

$\text{Var}(Y)=E[Y^2]-E[Y]^2$

```python
sum((y-mu_y)**2)/len(y)
```
